```python
# Meng-import modul-modul yang hendak digunakan.
import pandas as pd # Memasang label pada modul agar lebih mudah diketik.
from scipy.stats.mstats import gmean, hmean # Meng-import tool yang spesifik dari sebuah modul.
import matplotlib.pyplot as plt

# Menetapkan file yang akan dijadikan sumber data.
my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') # File dibaca oleh modul Pandas.

# Mendeklarasikan variabel.
a_mean = my_dataset.Sc.mean() # Mean = rata-rata.
g_mean = gmean(my_dataset['Sc']) 
h_mean = hmean(my_dataset['Sc'])

print ('-------') # Print menampilkan output dalam bentuk teks.
print ('Rerata Aritmetika')
print ("{0:.1f} [ppm]".format(a_mean)) # Perhitungan rata-ratanya. (?)
print ('-------')

print ('Rerata Geometrik')
print ("{0:.1f} [ppm]".format(g_mean))
print ('-------')

print ('Rerata Harmonik')
print ("{0:.1f} [ppm]".format(h_mean))
print ('-------')

# Grafik.
fig, ax = plt.subplots()
# Mengatur jenis grafik (Dalam hal ini, batang/histogram)
ax.hist(my_dataset.Sc, bins='auto', density=False, edgecolor='k', label='Rekaman Pengukuran', alpha=0.8) # Warna tepi, densitas?, serta label grafik pada legenda diatur di sini.
ax.axvline(a_mean, color="yellow", label='Rerata Aritmetika', linewidth=2) # Garis-garis tambahan untuk 3 variabel yang ditentukan sebelumnya.
ax.axvline(g_mean, color='blue',  label='Rerata Geometrik', linewidth=2)
ax.axvline(h_mean, color='#BA946B',  label='Rerata Harmonic', linewidth=2)
ax.set_xlabel('Sc [ppm]') # Label sumbu x.
ax.set_ylabel('Kemungkinan Tingkat Kepadatan') #Label sumbu y.
ax.legend() # Menambahkan legenda, terisi otomatis berdasarkan variabel-variabel dalam grafik.
ax.set_title("Saos's Bar Graph Numero Uno.", dict(fontname="times new roman", color="#BA946B"))

# Hasil awal sebelum code diubah. (Masih menggunakan data Zr)
'''
Output: 
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
# Ctt: Setelah diubah, output rerata geometrik dan harmonik menunjukkan 'nan' (Tidak bisa dihitung), sehingga tidak ditampilkan dalam grafik.
# Berdasarkan pesan error, ini mungkin terjadi karena data Sc yang digunakan ada yang bernilai negatif. (Baris 329 dan 354 dalam file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')
# Menentukan range tertentu, tidak mengambil keseluruhan data.
varian =  my_dataset['Nd'].var() # var (Variasi) mengukur kuadrat selisih data dengan rata-ratanya.
stddev =  my_dataset['Nd'].std() # std (Standar Deviasi) mengukur rata-rata jarak data dengan rata-rata data tersebut.

print ('-------')
print ('Variasi')
print("{0:.0f} [square ppm]".format(varian))
print ('-------')
print ('Standar Deviasi')
print("{0:.0f} [ppm]".format(stddev))
print ('-------')

fig, ax = plt.subplots()
ax.hist(my_dataset.Nd, bins= 8, density = True, edgecolor='k', label='Pengukuran Berdasar Waktu') 
ax.axvline(my_dataset['Nd'].mean()-stddev, color='#BA946B', label=r'Rerata - 1$\sigma$', linewidth=2)
ax.axvline(my_dataset['Nd'].mean()+stddev, color='blue', label=r'Rerata + 1$\sigma$', linewidth=2)
ax.axvspan(my_dataset['Nd'].mean()-stddev, my_dataset['Nd'].mean()+stddev, alpha=0.1, color='black', label=r'Rerata $\pm$ 1$\sigma$')
ax.set_xlabel('Nd [ppm]')
ax.set_ylabel('Mungkin padat')
ax.legend()
ax.set_title("Something something graphic yada yada saos etc etc", dict(fontname="times new roman", color="#BA946B"))
ax.text(70, 0.02, r'Masih ada ruang kosong.', dict(size=12), fontname="times new roman")

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
# Code yang satu ini mengalami error.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

q1 = np.percentile(my_dataset.Y, 37, method = 'midpoint') # Error pada 2 line ini.
q3 = np.percentile(my_dataset.Y, 71, method = 'midpoint') # Numpy tidak punya fungsi interpolation?
# Diperbaiki dengan mengganti interpolation menggunakan method.
iqr = q3 - q1 # Interquaritle Range (IQR)  

print ('-------')
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr))
print ('-------')

fig, ax = plt.subplots()
ax.set_title('Numero 6.')
ax.hist(my_dataset.Y, bins='auto', density=True, color='red', edgecolor='k', label='Sejarah') # bins = auto, jumlah batang ditentukan otomatis. 
ax.axvline(q1, color='#BA946B', label='Q1', linewidth=2)
ax.axvline(q3, color='blue', label='Q3', linewidth=2)
ax.axvspan(q1, q3, alpha=0.1, color='yellow', label='Interquaritle Range (IQR)')
ax.set_xlabel('Y [ppm]')
ax.set_ylabel('Memungkinkan Kepadatan dan sebagainya')
ax.legend()

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import numpy as np
from scipy.stats import skew # Skew menentukan apakah data terdistribusi dengan normal (Seimbang sisi kiri dan kanan, puncak di tengah), atau condong ke satu arah.

a_mean = my_dataset.Y.mean()
mid = my_dataset.Y.median()
hist, bin_edges = np.histogram(my_dataset['Y'], bins=20, density=True)
modus = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2
stddev = my_dataset['Y'].std()

kof1 = (a_mean - modus) / stddev # Rata-rata dikurang modus.
kof2 = 3 * (a_mean - mid) / stddev # 3 kali rata-rata dikurang median.
kof3 = skew(my_dataset['Y']) # Kecondongan dari data Y.

print ('PERHATIAN.')
print ("Koefisien Kecondongan ke-1 Pearson: {:.2f}".format(kof1))
print ("Momen Kecondongan ke-2 Pearson: {:.2f}".format(kof2))
print ("Koefisien Kecondongan Fisher-Pearson: {:.2f}".format(kof3))
print ('TERIMA KASIH ATAS PERHATIAN ANDA.')

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Kali ini menggunakan model grafik Boxplot.
my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

fig, ax = plt.subplots()
my_flierprops = dict(markerfacecolor="#c4ba89", markeredgecolor="#5c0003", marker='o') # 3 variabel yang diperlukan grafik Boxplot.
my_medianprops = dict(color="#1400ea", linewidth = 2)
my_boxprops = dict(facecolor="#000000", edgecolor="#8f0a0a")
ax.boxplot(my_dataset.Th, patch_artist = True, notch=True, flierprops=my_flierprops, medianprops=my_medianprops, boxprops=my_boxprops) 
ax.set_ylabel('Th [ppm]')
ax.set_xticks([1])
ax.set_xticklabels(['Semua Epoch'])
ax.set_title("Saos's Glorious BoxPlot Numero Cinqo.")
plt.show()
```
